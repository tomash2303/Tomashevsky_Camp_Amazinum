```python
import pandas as pd
import numpy as np
```

```python
def answer_one():
    energy = pd.read_excel("Energy Indicators (2).xls", skiprows=17, skipfooter=38)

    energy.columns = ["Delete 1", "Delete 2", "Country", "Energy Supply",
                      "Energy Supply per Capita", "% Renewable"]

    energy = energy.drop(['Delete 1', 'Delete 2'], axis=1)

    energy["Energy Supply"] = 1000000 * energy["Energy Supply"]

    energy = energy.replace(to_replace='\.+', value=np.NaN, regex=True)

    energy["Country"] = energy["Country"].replace(to_replace='Republic of Korea', 
                                                  value='South Korea', regex=True)
    energy["Country"] = energy["Country"].replace(to_replace='United States of America', 
                                                  value='United States', regex=True)
    energy["Country"] = energy["Country"].replace(to_replace='United Kingdom of Great Britain and Northern Ireland', 
                                                  value='United Kingdom', regex=True)
    energy["Country"] = energy["Country"].replace(to_replace='China, Hong Kong Special Administrative Region', 
                                                  value='Hong Kong', regex=True)

    energy = energy.replace(to_replace='[0-9]', value='', regex=True)
    energy = energy.replace(to_replace='\(.*\)', value='', regex=True)
    energy = energy.replace(to_replace='\s$', value='', regex=True)

    energy[energy['Country'] == 'Iran']

    ##########################################################################################

    gdp = pd.read_csv("gdp.csv", skiprows=4)
    #gdp['Country Name'] = 4

    gdp_replace_dict = {
        'Korea, Rep.': 'South Korea',
        'Iran, Islamic Rep.': 'Iran',
        'Hong Kong SAR, China': 'Hong Kong'
    }

    gdp['Country Name'] = gdp['Country Name'].replace(to_replace=gdp_replace_dict, regex=True)
    gdp = gdp.sort_values(by='Country Name')#.reset_index()
    gdp.sort_values(by='Country Name')[60:120]
    gdp[gdp['Country Name'] == 'United Kingdom']

    scim_en = pd.read_excel("scimagojr country rank 1996-2021 (1).xlsx")

    ############################################################################################

    energy_gdp = pd.merge(left=energy, right=gdp, left_on='Country', 
                          right_on='Country Name', how='inner')

    join_df = pd.merge(left=energy_gdp, right=scim_en, left_on='Country', 
                       right_on='Country', how='inner')

    join_df = join_df.set_index('Country')

    join_df = join_df[['Rank', 'Documents', 'Citable documents', 'Citations', 
                       'Self-citations', 'Citations per document', 'H index', 
                       'Energy Supply', 'Energy Supply per Capita', '% Renewable', 
                       '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']]

    join_df = join_df.sort_values(by='Rank').head(15)

    return join_df

answer_one()
```

```python
# What is the average GDP over the last 10 years for each country? 
# (exclude missing values from this calculation.)

# This function should return a Series named avgGDP with 15 countries 
# and their average GDP sorted in descending order.

def answer_two():
    top = answer_one()
    #df['sum'] = df[['2019', '2021']].apply(np.mean, axis=1)
    top['avg_gdp'] = top[['2006', '2007', '2008', '2009', '2010',
                          '2011', '2012', '2013', '2014', '2015']].apply(np.mean, axis=1)
    
    return top['avg_gdp'].sort_values(ascending=False)

print(answer_two())
    
```

```python
# By how much had the GDP changed over the 10 year span 
# for the country with the 6th largest average GDP?

#This function should return a single number.

def answer_three():
    top_df = answer_one()
    avg_gdp = answer_two()
    sixth_gdp_country = avg_gdp.index[5]
    #df.loc['Ukraine', '2021'] - df.loc['Ukraine', '2019']
    gdp_diff = top_df.loc[sixth_gdp_country, '2015'] - top_df.loc[sixth_gdp_country, '2006']
    return gdp_diff

print(answer_three())
    
```

```python
def answer_four():
    #df['ratio'] = df['2021'] / df['2019']
    top_df = answer_one()
    top_df['citations_ratio'] = top_df['Self-citations'] / top_df['Citations']
    citations_sorted_sr = top_df['citations_ratio'].sort_values(ascending=False)

    max_value = citations_sorted_sr.iloc[0]
    max_country = citations_sorted_sr.index[0]
 
    return (max_country, max_value)

print(answer_four())
    
```

```python
def answer_five():
    top_df = answer_one()
    top_df['population'] = top_df['Energy Supply'] / top_df['Energy Supply per Capita']
    populations_sorted_desc_sr = top_df['population'].sort_values(ascending=False)
    third_populous_country = populations_sorted_desc_sr.index[2]
    return third_populous_country

print(answer_five())
```

```python
def answer_six():
    top_df = answer_one()
    top_df['population'] = top_df['Energy Supply'] / top_df['Energy Supply per Capita']
    top_df['cit_docs_per_capita'] =  top_df['Citable documents'] / top_df['population']
    
    to_corr_df = top_df[['cit_docs_per_capita', 'Energy Supply per Capita']]
    coef_corr = to_corr_df.corr().iloc[0,1]
    return coef_corr
    
print(answer_six())
```

```python
def answer_seven():
    top_df = answer_one()
    top_df['population'] = top_df['Energy Supply'] / top_df['Energy Supply per Capita']
    continent_dict = {'China':'Asia', 
                  'United States':'North America', 
                  'Japan':'Asia', 
                  'United Kingdom':'Europe', 
                  'Russian Federation':'Europe', 
                  'Canada':'North America', 
                  'Germany':'Europe', 
                  'India':'Asia',
                  'France':'Europe', 
                  'South Korea':'Asia', 
                  'Italy':'Europe', 
                  'Spain':'Europe', 
                  'Iran':'Asia',
                  'Australia':'Australia', 
                  'Brazil':'South America'}
    df= pd.Series(data=continent_dict, name='continent').to_frame()
    top_df = pd.merge(left=top_df, right=df, left_index=True, 
                          right_index=True, how='inner')
    
    top_df_pop_cont = top_df[['population', 'continent']]
    
    #final_df = top_df_pop_cont.groupby('continent').apply(sum)
    final_df = top_df_pop_cont.groupby('continent').agg(['count', 'sum', 'mean', 'std'])
    return final_df
print(answer_seven())
    
    
```
